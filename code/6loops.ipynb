```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
multipliers = np.array((2, 3, 4, 5, 6, 9, 12, 18))
totals = multipliers*multipliers*2*3
```

```python
N_flip_triv = []
triv_std = []
for multiplier in multipliers:
    width  = multiplier*3
    height = multiplier*2
    rhos = pd.read_csv('longs/N_flippable' + str(width) + 'by' + str(height) + '.dat', header=None, skiprows=1).to_numpy() / (width*height)
    plt.plot(rhos[0,2:])
    N_flip_triv.append(np.average(rhos[0,2:]))
    triv_std.append(np.std(rhos[0,2:])/len(rhos[0,2:]))

# plt.ylim([1.55, 1.82])
plt.show()
```

```python
N_flip_nontriv = []
nontriv_std = []
for multiplier in multipliers:
    width  = multiplier*3
    height = multiplier*2
    rhos = pd.read_csv('longs/N_flippableNontrivial' + str(width) + 'by' + str(height) + '.dat', header=None, skiprows=1).to_numpy() / (width*height)
    plt.plot(rhos[0,2:])
    N_flip_nontriv.append(np.average(rhos[0,2:]))
    triv_std.append(np.std(rhos[0,2:])/len(rhos[0,2:]))

# plt.ylim([1.55, 1.82])
plt.show()
```

```python
plt.errorbar(multipliers, N_flip_triv, yerr=triv_std)
plt.errorbar(multipliers, N_flip_nontriv, yerr=nontriv_std)
plt.show()

plt.errorbar(multipliers, N_flip_triv*totals, yerr=triv_std)
plt.errorbar(multipliers, N_flip_nontriv*totals, yerr=nontriv_std)
plt.show()

plt.plot(np.log(multipliers), np.log(totals*N_flip_triv))
plt.plot(np.log(multipliers), np.log(totals*N_flip_nontriv))
plt.show()
```

```python
plt.plot(np.log(multipliers), np.log(totals*(np.array(N_flip_triv)-N_flip_nontriv)))
plt.plot(np.log(multipliers), .345*np.log(multipliers)+3.5)
plt.show()

plt.plot(np.log(multipliers), np.log(np.array(N_flip_triv)-N_flip_nontriv))
plt.plot(np.log(multipliers), -1.655*np.log(multipliers)+1.7)
plt.show()
```
