```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
multipliers = (2, 3, 4, 5, 6, 9, 12)
for multiplier in multipliers:
    # print('data/N_flippable' + str(width) + 'by' + str(height) + '.dat')
    width  = multiplier*3
    height = multiplier*2
    rhos = pd.read_csv('data/N_flippable' + str(width) + 'by' + str(height) + '.dat').to_numpy() / (width*height)
    rho_bar = np.average(rhos, axis=0)
    rho_std = np.std(rhos, axis=0) / np.sqrt(np.shape(rhos)[0])
    # plt.errorbar(range(len(rho_bar)), rho_bar, yerr=rho_std, errorevery= 3)
    plt.plot(rho_bar)

plt.ylim([1.55, 1.82])
plt.show()
```

```python
rho_bars = []
for multiplier in multipliers:
    width  = multiplier*3
    height = multiplier*2
    rhos = pd.read_csv('data/N_flippable' + str(width) + 'by' + str(height) + '.dat').to_numpy()
    rhos = rhos / (width*height)
    rho_bar = np.average(rhos, axis=0)
    rho_bars.append(np.average(rho_bar[300:600]))
    
plt.plot(np.log(multipliers), np.log(rho_bars))
plt.show()
```

```python
for multiplier in range(2,5):
    # print('data/N_flippable' + str(width) + 'by' + str(height) + '.dat')
    width  = multiplier*3
    height = multiplier*2
    rhos1 = pd.read_csv('data/N_flippable' + str(width) + 'by' + str(height) + '.dat').to_numpy()
    rhos1_bar = np.average(rhos1, axis=0)
    rhos2 = pd.read_csv('data/N_flippableNontrivial' + str(width) + 'by' + str(height) + '.dat').to_numpy()
    rhos2_bar = np.average(rhos2, axis=0)
    delta = rhos1_bar - rhos2_bar
    plt.plot(delta)

plt.ylim([-2,2])
plt.show()
```

```python
deltas = []
for multiplier in multipliers:
    width  = multiplier*3
    height = multiplier*2
    rhos2 = pd.read_csv('data/N_flippableNontrivial' + str(width) + 'by' + str(height) + '.dat').to_numpy()
    rhos2_bar = np.average(rhos2, axis=0)
    rhos1 = pd.read_csv('data/N_flippable' + str(width) + 'by' + str(height) + '.dat').to_numpy()
    rhos1_bar = np.average(rhos1, axis=0)
    rho_diff = rhos1_bar - rhos2_bar
    deltas.append(np.average(rho_diff[150:200]))
    
plt.plot(multipliers, deltas)
# plt.plot(np.log(multipliers), np.log(np.abs(deltas)))
plt.show()
```
